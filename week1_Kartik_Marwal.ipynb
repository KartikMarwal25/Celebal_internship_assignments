````python
A = np.array([[4, 1], [2, 3]], dtype=float)

eigenvalues, eigenvectors = np.linalg.eig(A)

print(f"Eigenvalues: {eigenvalues}")
print(f"Eigenvectors (columns):\n{eigenvectors}")

# Verify Av = λv for each eigenpair
# Your Code Here
for i in range(len(eigenvalues)):
    lambda_val = eigenvalues[i]
    v_vec = eigenvectors[:, i]
    Av = A @ v_vec
    lambda_v = lambda_val * v_vec
    print(f"\nVerifying for eigenvalue {lambda_val:.2f} and eigenvector {v_vec}:")
    print(f"A @ v = {Av}")
    print(f"λ * v = {lambda_v}")
    assert np.allclose(Av, lambda_v), f"Verification failed for eigenpair {i}"

# Plot
# Your Code Here
fig, ax = plt.subplots(figsize=(6, 6))

# Plot scaled eigenvectors
ax.quiver(0, 0, eigenvectors[0, 0] * eigenvalues[0], eigenvectors[1, 0] * eigenvalues[0],
          angles='xy', scale_units='xy', scale=1, color='r', width=0.01, label='Eigenvector 1 (scaled)')
ax.quiver(0, 0, eigenvectors[0, 1] * eigenvalues[1], eigenvectors[1, 1] * eigenvalues[1],
          angles='xy', scale_units='xy', scale=1, color='b', width=0.01, label='Eigenvector 2 (scaled)')

ax.set_xlim(-max(abs(eigenvalues)) * 1.5, max(abs(eigenvalues)) * 1.5)
ax.set_ylim(-max(abs(eigenvalues)) * 1.5, max(abs(eigenvalues)) * 1.5)
ax.set_xlabel('x-axis')
ax.set_ylabel('y-axis')
ax.set_title('Eigenvectors scaled by Eigenvalues')
ax.axhline(0, color='k', linewidth=0.5)
ax.axvline(0, color='k', linewidth=0.5)
ax.grid(True, linestyle='--', alpha=0.6)
ax.legend()
plt.tight_layout()
plt.show()

print("4.3 passed ✓")
````

**Your Geometric Explanation (4.3):**

When a matrix stretches a vector, it means that after applying the transformation represented by the matrix, the vector keeps pointing in the same direction but its magnitude changes. In other words, the vector may become longer or shorter without changing its orientation. Such a vector is called an eigenvector. The factor by which the vector is stretched or shrunk is known as the eigenvalue. Therefore, eigenvectors represent the special directions that remain unchanged under the matrix transformation, while eigenvalues indicate the amount of scaling along those directions.

### 5.4 Distribution Testing & Stationarity

1. Generate two samples: `s1` ~ Normal(0, 1), `s2` ~ Exponential(1), both n=200 (seed=0).
2. Run the **Kolmogorov-Smirnov test** (`stats.kstest`) comparing each against a standard normal CDF.
3. Generate a non-stationary time series (linear trend + noise). Run the **Augmented Dickey-Fuller test** (`adfuller` from `statsmodels`) and interpret the result.
4. Difference the series once and re-run ADF to confirm stationarity.

````python
from statsmodels.tsa.stattools import adfuller

np.random.seed(0)
s1 = np.random.normal(0, 1, 200)
s2 = np.random.exponential(1, 200)

# KS test — compare each against N(0,1)
ks_s1 = stats.kstest(s1, 'norm', args=(0, 1))  # YOUR CODE HERE
ks_s2 = stats.kstest(s2, 'norm', args=(0, 1))  # YOUR CODE HERE
print(f"KS test s1 (normal):      stat={ks_s1.statistic:.4f}, p={ks_s1.pvalue:.4f}")
print(f"KS test s2 (exponential): stat={ks_s2.statistic:.4f}, p={ks_s2.pvalue:.4f}")

# ADF test on non-stationary series
t = np.arange(200)
ts = 0.05 * t + np.random.normal(0, 1, 200)

adf_result = adfuller(ts)  # YOUR CODE HERE — adfuller(ts)
print(f"\nADF statistic: {adf_result[0]:.4f}")
print(f"ADF p-value:   {adf_result[1]:.4f}")
print(f"Stationary?    {adf_result[1] < 0.05}")

# Difference once and re-run
ts_diff = np.diff(ts)  # YOUR CODE HERE — np.diff
adf_diff = adfuller(ts_diff)  # YOUR CODE HERE
print(f"\nAfter differencing — p-value: {adf_diff[1]:.4f}, Stationary? {adf_diff[1] < 0.05}")
````

````python
def compute_psi(expected, actual, bins=10):
    """
    PSI = sum((actual_pct - expected_pct) * ln(actual_pct / expected_pct))
    Add epsilon=1e-10 to avoid log(0).
    """
    # Determine the bin edges across both distributions
    all_data = np.concatenate([expected, actual])
    hist_counts, bin_edges = np.histogram(all_data, bins=bins)

    # Calculate counts in each bin for expected and actual distributions
    expected_bin_counts, _ = np.histogram(expected, bins=bin_edges)
    actual_bin_counts, _ = np.histogram(actual, bins=bin_edges)

    # Calculate percentages, adding epsilon to avoid division by zero
    epsilon = 1e-10
    expected_percentages = (expected_bin_counts / len(expected)) + epsilon
    actual_percentages = (actual_bin_counts / len(actual)) + epsilon

    # Calculate PSI for each bin and sum them
    psi_per_bin = (actual_percentages - expected_percentages) * np.log(actual_percentages / expected_percentages)
    psi_value = np.sum(psi_per_bin)
    return psi_value

np.random.seed(1)
train_dist = np.random.normal(50, 10, 1000)    # training distribution
drift_dist = np.random.normal(65, 12, 1000)    # shifted production distribution

psi_value = compute_psi(train_dist, drift_dist)
print(f"PSI: {psi_value:.4f}")
print(f"Shift severity: {'Major' if psi_value > 0.2 else 'Minor' if psi_value > 0.1 else 'Stable'}")

# Plot both distributions
fig, ax = plt.subplots()
sns.histplot(train_dist, kde=True, color='blue', label='Training Distribution', ax=ax, stat='density', alpha=0.5)
sns.histplot(drift_dist, kde=True, color='red', label='Drift Distribution', ax=ax, stat='density', alpha=0.5)
ax.set_xlabel('Value')
ax.set_ylabel('Density')
ax.legend(); ax.set_title('Distribution Shift Visualization')
plt.tight_layout(); plt.show()
````

**Concept Drift vs Covariate Drift & PSI Thresholds (5.5):**

- **Concept drift:** Occurs when the relationship between the input features and the target variable changes over time. This means the underlying concept that the model is trying to learn has evolved, and the model's predictions may no longer be accurate even if the input data distribution remains the same.
- **Covariate drift:** Occurs when the distribution of the input features (covariates) changes over time, while the relationship between the inputs and the target variable remains consistent. The input data shifts, but the underlying concept the model learned is still valid for the *new* distribution.
- **PSI < 0.1:** Indicates **no significant shift**; the distributions are considered stable. Model performance is likely to be consistent.
- **PSI 0.1–0.2:** Indicates a **minor shift**; this level might warrant further investigation or monitoring, but not necessarily immediate action. The model's performance might be slightly degraded.
- **PSI > 0.2:** Indicates a **major shift**; this typically suggests that the model's performance is likely degrading significantly and requires immediate attention, such as retraining or thorough investigation into the cause of the drift.
- **Retraining trigger example:** A common trigger for retraining a model would be when the Population Stability Index (PSI) for critical input features or the model's output predictions consistently exceeds the **major shift threshold (e.g., PSI > 0.2)** for a predefined period (e.g., three consecutive days). This signals that the model's operating environment has changed enough to likely impact its accuracy.

A bag contains: 4 red, 3 blue, 3 green marbles.

1. Define the sample space and compute P(red), P(blue), P(green).
2. Two marbles drawn **without replacement**. Compute:
   - P(first=red, second=blue) — **joint probability**
   - P(second=blue | first=red) — **conditional probability**
3. Are the two draws **independent**? Show mathematically.

````python
total = 10
red, blue, green = 4, 3, 3

p_red   = red / total
p_blue  = blue / total
p_green = green / total

# Joint: P(red first AND blue second)
p_red_then_blue = (red / total) * (blue / (total - 1))

# Conditional: P(blue second | red first)
p_blue_given_red = blue / (total - 1)

# Independence check: are P(B|A) == P(B) ?
independent = p_blue_given_red == p_blue

print(f"P(red)={p_red:.3f}, P(blue)={p_blue:.3f}, P(green)={p_green:.3f}")
print(f"P(red,blue): {p_red_then_blue:.4f}")
print(f"P(blue|red): {p_blue_given_red:.4f}")
print(f"P(blue):     {p_blue:.4f}")
print(f"Independent? {independent}")
````

### 6.3 Bayes' Theorem

**Scenario — Spam filter:**
- P(Spam) = 0.30 (prior)
- P("free" | Spam) = 0.80 (likelihood)
- P("free" | Not Spam) = 0.05

1. Compute P(Spam | "free") using Bayes' theorem.
2. Implement a simple `naive_bayes_predict(prior_spam, p_word_given_spam, p_word_given_ham)` that returns the posterior P(Spam | word).
3. In the markdown cell, map each term — **prior, likelihood, evidence, posterior** — to the spam example.

````python
p_spam            = 0.30
p_free_given_spam = 0.80
p_free_given_ham  = 0.05
p_ham             = 1 - p_spam

# P(free) — law of total probability
p_free = (p_free_given_spam * p_spam) + (p_free_given_ham * p_ham)

# Bayes
p_spam_given_free = (p_free_given_spam * p_spam) / p_free

print(f"P('free'): {p_free:.4f}")
print(f"P(Spam | 'free'): {p_spam_given_free:.4f}")

def naive_bayes_predict(prior_spam, p_word_given_spam, p_word_given_ham):
    # P(Not Spam)
    prior_ham = 1 - prior_spam
    # P(word) using law of total probability
    p_word = (p_word_given_spam * prior_spam) + (p_word_given_ham * prior_ham)
    # Bayes' Theorem
    posterior_spam = (p_word_given_spam * prior_spam) / p_word
    return posterior_spam

pred = naive_bayes_predict(0.30, 0.80, 0.05)
assert abs(pred - p_spam_given_free) < 1e-9
print(f"naive_bayes_predict: {pred:.4f}")
print("6.3 passed ✓")
````

**Bayes Term Mapping (6.3):**

| Term | Formula | In this example |
|------|---------|-----------------|
| Prior | P(Spam) | P(Spam) = 0.30 (Initial belief about spam) |
| Likelihood | P(word \| Spam) | P("free" \| Spam) = 0.80 (Probability of seeing "free" given it's spam) |
| Evidence | P(word) | P("free") = 0.275 (Overall probability of seeing "free") |
| Posterior | P(Spam \| word) | P(Spam \| "free") = 0.8727 (Updated belief about spam after seeing "free") |

### 6.4 Central Limit Theorem

1. Define a population following an **exponential distribution** (λ=1, size=100,000).
2. Repeatedly draw samples of size `n=30` and record their means — do this 5,000 times.
3. Plot the histogram of sample means.
4. Overlay the **theoretical normal distribution** predicted by the CLT: μ = population mean, σ = population_std / sqrt(n).
5. Run a KS test to confirm the distribution of sample means is approximately normal.

````python
np.random.seed(7)
population = np.random.exponential(scale=1.0, size=100_000)

n_samples = 5000
sample_size = 30

sample_means = np.array([np.mean(np.random.choice(population, size=sample_size, replace=False)) for _ in range(n_samples)])

pop_mean = population.mean()
pop_std  = population.std()
clt_std  = pop_std / np.sqrt(sample_size)

print(f"Population mean: {pop_mean:.4f}, Population std: {pop_std:.4f}")
print(f"Sample means mean: {sample_means.mean():.4f}, std: {sample_means.std():.4f}")
print(f"CLT predicted std: {clt_std:.4f}")

# Plot
fig, ax = plt.subplots()
sns.histplot(sample_means, kde=True, stat='density', ax=ax, color='blue', label='Sample Means Distribution')

x_norm = np.linspace(min(sample_means), max(sample_means), 100)
ax.plot(x_norm, stats.norm.pdf(x_norm, loc=pop_mean, scale=clt_std), color='red', linestyle='--', label='Theoretical Normal (CLT)')

ax.set_title('CLT: Distribution of Sample Means')
ax.set_xlabel('Sample Mean')
ax.set_ylabel('Density')
ax.legend()
plt.tight_layout(); plt.show()

# KS test
ks_result = stats.kstest(sample_means, 'norm', args=(pop_mean, clt_std))
print(f"KS test p-value: {ks_result.pvalue:.4f} → Approximately normal? {ks_result.pvalue > 0.05}")
````

**CLT Reflection (6.4):**

 The Central Limit Theorem (CLT) states that if we take a large number of samples from any population, then the distribution of the sample means becomes approximately normal, even if the original data is not normally distributed.

This theorem is very important in statistics and machine learning because many statistical methods assume normal distribution. CLT makes it possible to use those methods on real-world data. It helps in hypothesis testing, confidence intervals, prediction, and making conclusions about a population using sample data.
